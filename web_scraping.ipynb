```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get("https://www.worlddata.info/developing-countries.php").text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('td')[100]
```

```python
c = []
for i in soup.find_all('td'):
    c.append(i.text.strip())
```

```python
c
```

```python
country = []
for i in range(796):
    if i%5 == 0:
        country.append(c[i])
    if c[i] == 'Zimbabwe':
        break
```

```python
print(country)
len(country)
```

```python
population = []
for i in range(796):
    if i%5 == 1:
        population.append(c[i])
    if c[i-1] == 'Zimbabwe':
        break

print(population)
len(population)
```

```python
GNI_per_capita = []
for i in range(796):
    if i%5 == 2:
        GNI_per_capita.append(c[i])
    if c[i-2] == 'Zimbabwe':
        break

print(GNI_per_capita)
len(GNI_per_capita)
```

```python
human_development_index = []
for i in range(796):
    if i%5 == 3:
        human_development_index.append(c[i])
    if c[i-3] == 'Zimbabwe':
        break

print(human_development_index)
len(human_development_index)
```

```python
human_asset_index = []
for i in range(796):
    if i%5 == 4:
        human_asset_index.append(c[i])
    if c[i-4] == 'Zimbabwe':
        break

print(human_asset_index)
len(human_asset_index)
```

```python
df = pd.DataFrame()
```

```python
df['COUNTRY'] = country
df['POPULATION(MILLION)'] = population
df['GNI_PER_CAPITA(USD)'] = GNI_per_capita
df['HDI'] = human_development_index
df['HAI'] = human_asset_index
```

```python
df
```

```python
df.info()
```

```python
df['HAI'] = pd.to_numeric(df['HAI'], errors='coerce')
df['HDI'] = pd.to_numeric(df['HDI'], errors='coerce')
```

```python
for i in range(152):
    df['POPULATION(MILLION)'][i] = df['POPULATION(MILLION)'][i].split()[0]
```

```python
for i in range(152):
    df['GNI_PER_CAPITA(USD)'][i] = df['GNI_PER_CAPITA(USD)'][i].split()[0]
```

```python
df.info()
```

```python
df['GNI_PER_CAPITA(USD)'] = df['GNI_PER_CAPITA(USD)'].str.replace(',', '').astype(float)
```

```python
df['POPULATION(MILLION)'] = df['POPULATION(MILLION)'].str.replace(',', '').astype(float)
```
